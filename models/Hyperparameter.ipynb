```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import time

# Custom RandomForestClassifier function
def custom_RandomForestClassifier(n_estimators=100, criterion='gini', max_depth=None, min_samples_split=2, 
                                  min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_features='sqrt', 
                                  max_leaf_nodes=None, min_impurity_decrease=0.0, bootstrap=True, oob_score=False, 
                                  n_jobs=None, random_state=None, verbose=0, warm_start=False, class_weight=None, 
                                  ccp_alpha=0.0, max_samples=None):
    
    # Instantiate the RandomForestClassifier model with the provided parameters
    model = RandomForestClassifier(
        n_estimators=n_estimators,
        criterion=criterion,
        max_depth=max_depth,
        min_samples_split=min_samples_split,
        min_samples_leaf=min_samples_leaf,
        min_weight_fraction_leaf=min_weight_fraction_leaf,
        max_features=max_features,
        max_leaf_nodes=max_leaf_nodes,
        min_impurity_decrease=min_impurity_decrease,
        bootstrap=bootstrap,
        oob_score=oob_score,
        n_jobs=n_jobs,
        random_state=random_state,
        verbose=verbose,
        warm_start=warm_start,
        class_weight=class_weight,
        ccp_alpha=ccp_alpha,
        max_samples=max_samples
    )
    
    return model

# Example usage
def evaluate_random_forest(X_train, X_test, y_train, y_test, name="RandomForest"):
    # Use the custom RandomForestClassifier function to get the model
    model = custom_RandomForestClassifier(n_estimators=100,max_depth=10, max_features='sqrt')
    
    start_time = time.time()
    print(f"Training {name}...")
    
    # Fit the model
    model.fit(X_train, y_train)
    
    # Make predictions
    y_pred = model.predict(X_test)
    
    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    
    # Calculate time taken
    time_consumed = time.time() - start_time
    
    print(f"{name} Accuracy: {round(accuracy, 2)} Time: {time_consumed}")

# Example call with train and test data (X_train, X_test, y_train, y_test should be predefined)
# evaluate_random_forest(X_train, X_test, y_train, y_test)
```
